```python
#Question->1
# part a

import numpy as np

arr = np.array([1, 2, 3, 6, 4, 5])
reversed_arr = arr[::-1]
print(reversed_arr)
#---------------------------------------------------------------

# part b

array1 = np.array([[1, 2, 3], [2, 4, 5], [1, 2, 3]])

flat_array1 = array1.flatten() #using flatten()

flat_array2 = array1.ravel() #using ravel()

print(flat_array1)
print(flat_array2)
#----------------------------------------------------------------

# part c

arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[1, 2], [3, 4]])

are_equal = np.array_equal(arr1, arr2)
print(are_equal)
#----------------------------------------------------------------

# part d (i)

x = np.array([1, 2, 3, 4, 5, 1, 2, 1, 1, 1])
most_frequent = np.bincount(x).argmax()
indices = np.where(x == most_frequent)
print(f"Most frequent value: {most_frequent}, Indices: {indices}")

# part d (ii)

y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
most_frequent = np.bincount(y).argmax()
indices = np.where(y == most_frequent)
print(f"Most frequent value: {most_frequent}, Indices: {indices}")
#------------------------------------------------------------------

# part e

gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

# i. Sum of all elements
sum_all = gfg.sum()

# ii. Sum of elements row-wise
sum_row_wise = gfg.sum(axis=1)

# iii. Sum of elements column-wise
sum_column_wise = gfg.sum(axis=0)

print(f"Sum of all elements: {sum_all}")
print(f"Sum row-wise: {sum_row_wise}")
print(f"Sum column-wise: {sum_column_wise}")
#-------------------------------------------------------------------

#part f

n_array = np.array([[55, 25, 15], [30, 44, 2], [11, 45, 77]])

# i. Sum of diagonal elements
diagonal_sum = np.trace(n_array)

# ii and iii Eigenvalues and eigenvectors of the matrix
eigen_values, eigen_vectors = np.linalg.eig(n_array)

# iv. Inverse of the matrix
inverse_matrix = np.linalg.inv(n_array)

# v. Determinant of the matrix
determinant = np.linalg.det(n_array)

print(f"Diagonal Sum: {diagonal_sum}")
print(f"Eigenvalues: {eigen_values}")
print(f"Eigenvectors: \n{eigen_vectors}")
print(f"Inverse: \n{inverse_matrix}")
print(f"Determinant: {determinant}")
#----------------------------------------------------------------------

# part g

# (i)
p = np.array([[1, 2], [2, 3]])
q = np.array([[4, 5], [6, 7]])

# Matrix multiplication
result1 = np.dot(p, q)

# Covariance
covariance1 = np.cov(p, q, rowvar=False)

print(f"Matrix multiplication result:\n{result1}")
print(f"Covariance:\n{covariance1}")

# (ii)

p = np.array([[1, 2], [2, 3], [4, 5]])
q = np.array([[4, 5, 1], [6, 7, 2]])

# Matrix multiplication
result2 = np.dot(p, q)

# Covariance
covariance2 = np.cov(p, q, rowvar=False)

print(f"Matrix multiplication result:\n{result2}")
print(f"Covariance:\n{covariance2}")
#--------------------------------------------------------------------

# part h

x = np.array([[2, 3, 4], [3, 2, 9]])
y = np.array([[1, 5, 0], [5, 10, 3]])

# Inner product
inner_product = np.inner(x, y)

# Outer product
outer_product = np.outer(x, y)

# Cartesian product
cartesian_product = np.transpose([np.tile(x.ravel(), len(y.ravel())), np.repeat(y.ravel(), len(x.ravel()))])

print(f"Inner product:\n{inner_product}")
print(f"Outer product:\n{outer_product}")
print(f"Cartesian product:\n{cartesian_product}")
#--------------------------------------------------------------------------------------------------------------------
```

```python
#Question ->2
# part a

# (i)
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

absolute_value = np.abs(array)
print(f"Element-wise absolute value:\n{absolute_value}")

# (ii)
# for flattened array
flattened = array.flatten()
percentile_25 = np.percentile(flattened, 25)
percentile_50 = np.percentile(flattened, 50)
percentile_75 = np.percentile(flattened, 75)

print(f"25th percentile (flattened): {percentile_25}")
print(f"50th percentile (flattened): {percentile_50}")
print(f"75th percentile (flattened): {percentile_75}")

# for each column
percentile_25_col = np.percentile(array, 25, axis=0)
percentile_50_col = np.percentile(array, 50, axis=0)
percentile_75_col = np.percentile(array, 75, axis=0)

print(f"25th percentile (column-wise): {percentile_25_col}")
print(f"50th percentile (column-wise): {percentile_50_col}")
print(f"75th percentile (column-wise): {percentile_75_col}")

# for each row
percentile_25_row = np.percentile(array, 25, axis=1)
percentile_50_row = np.percentile(array, 50, axis=1)
percentile_75_row = np.percentile(array, 75, axis=1)

print(f"25th percentile (row-wise): {percentile_25_row}")
print(f"50th percentile (row-wise): {percentile_50_row}")
print(f"75th percentile (row-wise): {percentile_75_row}")

# (iii)
# for flattened array
mean_flat = np.mean(flattened)
median_flat = np.median(flattened)
std_flat = np.std(flattened)

print(f"Mean (flattened): {mean_flat}")
print(f"Median (flattened): {median_flat}")
print(f"Standard Deviation (flattened): {std_flat}")

# for each column
mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

print(f"Mean (column-wise): {mean_col}")
print(f"Median (column-wise): {median_col}")
print(f"Standard Deviation (column-wise): {std_col}")

# for each row
mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print(f"Mean (row-wise): {mean_row}")
print(f"Median (row-wise): {median_row}")
print(f"Standard Deviation (row-wise): {std_row}")
#-------------------------------------------------------------------------------

# part b

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)

print(f"Floor values: {floor_values}")
print(f"Ceiling values: {ceiling_values}")
print(f"Truncated values: {truncated_values}")

rounded_values = np.round(a)
print(f"Rounded values: {rounded_values}")
#-------------------------------------------------------------------------------
```

```python
# Question->3
# part a
# (i)
import numpy as np

array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(array)
print(f"Sorted array: {sorted_array}")

# (ii)
sorted_indices = np.argsort(array)
print(f"Indices of sorted array: {sorted_indices}")

# (iii)
smallest_elements = np.sort(array)[:4]
print(f"4 smallest elements: {smallest_elements}")

# (iv)
largest_elements = np.sort(array)[-5:]
print(f"5 largest elements: {largest_elements}")
#-------------------------------------------------------------------------------

# part b
# (i)
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

integer_elements = array[array == np.floor(array)]
print(f"Integer elements only: {integer_elements}")

#(ii)
float_elements = array[array != np.floor(array)]
print(f"Float elements only: {float_elements}")
#-------------------------------------------------------------------------------


```
